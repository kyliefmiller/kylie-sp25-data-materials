```python
utah_map = folium.Map(location=[40, -111], zoom_start=6)
utah_map
```

```python
folium.Marker(location=[38.41, -112.339], popup="Adamsville Post Office").add_to(utah_map)
utah_map
```

```python
def create_map_markers(row, map_name):
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=row['place']).add_to(map_name)
```

```python
import pandas as pd
post_df = pd.read_csv('1871-utah-postmaster-salaries.csv', delimiter=",")
post_df.head()
```

```python
post_df.apply(create_map_markers, map_name=utah_map, axis='columns')
utah_map
```
