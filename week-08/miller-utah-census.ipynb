```python
def extend_marriage(letter):
    if letter == 'm':
        return 'married'
    elif letter == 'w':
        return 'widow'
    elif letter == 's':
        return 'single'

extend_marriage('m')
```

```python
utah_df['marrystatus'] = utah_df['marrystatus'].apply(extend_marriage)
```

```python
utah_df
```

```python
def birthyear_calc(age):
    birthyear = 1880-age
    return birthyear

birthyear_calc(18)
```

```python
utah_df['age'].apply(birthyear_calc)
```

```python
utah_df['birthyear'] = utah_df['age'].apply(birthyear_calc)
```

```python
utah_df
```

```python
utah_df['birthyear']=utah_df['birthyear'].astype('int64')
```

```python
birthsbyyear = utah_df.groupby('birthyear')['key'].count()
birthsbyyear
```
