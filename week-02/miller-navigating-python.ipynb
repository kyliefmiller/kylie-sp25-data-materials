Access the Lucy Parsons Wikipedia page (# Python 3)

```python
import urllib.request

fid=urllib.request.urlopen('https://en.wikipedia.org/wiki/Lucy_Parsons')

parsons_webpage=fid.read().decode('utf-8')
```

Print the title of the Wikipedia page (# extract webpage title)

```python
for line in parsons_webpage.split('\n'):

    if '<title>' in line:

        pagetitle=line.split('<title>')[1].split('</title>')[0]

        print(pagetitle)
```

Download a copy of the page named parsons.html (# save as local file 'webpage.html')

```python
import urllib.request

urllib.request.urlretrieve('https://en.wikipedia.org/wiki/Lucy_Parsons', 'parsons.html')
```

View the downloaded HTML file (# read local file)

```python
for line in open('parsons.html'):

    print(line.strip())
```
